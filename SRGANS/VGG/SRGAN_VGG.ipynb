````python
class ImageLoader:
  def __init__(self, root='', batch_size=4, mode = "train"):
    self.__dir = os.path.join(root, "immagini", f"DIV2K_{mode}")
    self.__images = os.listdir(os.path.join(self.__dir, "x1"))
    shuffle(self.__images)
    self.__nimages = len(self.__images)
    self.__batch_size = batch_size
    self.__index = 0
    self.__len = self.__nimages // self.__batch_size + (1 if self.__nimages % self.__batch_size != 0 else 0)

  def reset(self):
    shuffle(self.__images)
    self.__index = 0
  def batch_size(self):
    return self.__batch_size
  def nimages(self):
    return self.__nimages

  def __len__(self):
    return self.__len
  def __iter__(self):
    return self

  def __get_item__(self, index):
    original_x1 = cv2.cvtColor(cv2.imread(os.path.join(self.__dir, "x1", self.__images[index])), cv2.COLOR_BGR2RGB)

    rh, rw = (randint(0, original_x1.shape[0]-256), randint(0, original_x1.shape[1]-256))
    original_x1 = original_x1[rh:rh+256, rw:rw+256]

    rotate, rotation_choice = (randint(0, 1) == 1, [cv2.ROTATE_90_CLOCKWISE, cv2.ROTATE_180, cv2.ROTATE_90_COUNTERCLOCKWISE][randint(0, 2)])
    flip, flip_choice = (randint(0, 1) == 1, randint(0, 1))

    if rotate:
      original_x1 = cv2.rotate(original_x1, rotation_choice)
    if flip:
      original_x1 = cv2.flip(original_x1, flip_choice)

    lowres = cv2.resize(original_x1, (64,64), interpolation=cv2.INTER_CUBIC)

    hr = np.array(original_x1, dtype=np.float32)/255.
    lr = np.array(lowres, dtype=np.float32)/255.
    return hr, lr

  def __next__(self):
    if self.__index == self.__len:
      self.reset()
      raise StopIteration
    batch_hr = []
    batch_lr = []
    for i in range(self.__batch_size):
      current_index = self.__index*self.__batch_size + i
      if current_index >= self.__nimages:
        break
      hr, lr = self.__get_item__(current_index)
      batch_hr.append(hr)
      batch_lr.append(lr)
    self.__index+=1
    return tf.stack(batch_hr), tf.stack(batch_lr)
````

````python
def ResBlock(filters=64, kernel_size=3, momentum=0.8):
  def __call__(inputs):
    x = tf.keras.layers.Conv2D(filters=filters, kernel_size=kernel_size, padding="same", use_bias=False)(inputs)
    x = tf.keras.layers.BatchNormalization(momentum=momentum)(x)
    x = tf.keras.layers.PReLU(shared_axes=[1,2], alpha_initializer=tf.keras.initializers.constant(2e-2))(x)
    x = tf.keras.layers.Conv2D(filters=filters, kernel_size=kernel_size, padding="same", use_bias=False)(x)
    x = tf.keras.layers.BatchNormalization(momentum=momentum)(x)
    x = tf.keras.layers.Add()([x, inputs])
    return x
  return __call__

def Upscale2D(scale_factor: int = 2, momentum=0.8):
  def __call__(inputs):
    x = tf.keras.layers.Conv2D(filters=256, kernel_size=3, padding="same")(inputs)
    x = tf.keras.layers.UpSampling2D(scale_factor)(x)
    x = tf.keras.layers.PReLU(shared_axes=[1,2], alpha_initializer=tf.keras.initializers.constant(2e-3))(x)
    return x
  return __call__

def Generator(input_shape=(None,None,3), n_residuals: int = 16, scale_factor: int = 4):
  n_residuals = n_residuals if n_residuals != None else scale_factor**2
  scale_step = int(np.log2(scale_factor))

  # input layer
  inlayer = tf.keras.Input(shape=input_shape)
  x = tf.keras.layers.Conv2D(filters=64, kernel_size=9, padding="same")(inlayer)
  x = tf.keras.layers.PReLU(shared_axes=[1,2], alpha_initializer=tf.keras.initializers.constant(2e-2))(x)

  # SRResNet
  checkpoint = x
  for _ in range(n_residuals):
    x = ResBlock()(x)
  x = tf.keras.layers.Conv2D(filters=64, kernel_size=3, padding="same", use_bias=False)(x)
  x = tf.keras.layers.BatchNormalization(momentum=0.8)(x)
  x = tf.keras.layers.Add()([checkpoint, x])

  # UpScaler
  for _ in range(scale_step):
    x = Upscale2D(scale_step)(x)

  # output
  x = tf.keras.layers.Conv2D(filters=3, kernel_size=9, padding="same", activation="sigmoid")(x)
  return tf.keras.Model(inputs=[inlayer], outputs=[x], name="Genny")
````

````python
def FeatureBlock(filters, kernel_size=3, momentum=0.8, alpha=0.2):
  def __call__(inputs):
    x = tf.keras.layers.Conv2D(filters=filters, kernel_size=kernel_size, use_bias=False, padding="same")(inputs)
    x = tf.keras.layers.BatchNormalization(momentum=momentum)(x)
    x = tf.keras.layers.LeakyReLU(alpha=alpha)(x)
    x = tf.keras.layers.Conv2D(filters=filters, kernel_size=kernel_size, use_bias=False, strides=2)(x)
    x = tf.keras.layers.BatchNormalization(momentum=momentum)(x)
    x = tf.keras.layers.LeakyReLU(alpha=alpha)(x)
    return x
  return __call__

def Discriminator(input_shape=(256,256,3)):
  inlayer = tf.keras.Input(shape=input_shape)

  # input layer
  x = tf.keras.layers.Conv2D(filters=64, kernel_size=3, padding="same")(inlayer)
  x = tf.keras.layers.LeakyReLU(alpha=0.2)(x)
  x = tf.keras.layers.Conv2D(filters=64, kernel_size=3, strides=2)(x)
  x = tf.keras.layers.BatchNormalization(momentum=0.8)(x)
  x = tf.keras.layers.LeakyReLU(alpha=0.2)(x)

  # features
  for n_channels in [128,256,512]:
    x = FeatureBlock(filters=n_channels)(x)

  # classifier
  x = tf.keras.layers.Flatten()(x)
  x = tf.keras.layers.Dense(1024)(x)
  x = tf.keras.layers.LeakyReLU(alpha=0.2)(x)
  x = tf.keras.layers.Dense(1, activation="sigmoid")(x) # 0 = Fake; 1 = Real;

  return tf.keras.Model(inputs=[inlayer], outputs=[x], name="Discostu")
````

````python
def psnr_metric(y_true, y_pred):
  return tf.image.psnr(y_true, y_pred, max_val=1.)

def ssim_metric(y_true, y_pred):
  return tf.image.ssim(y_true, y_pred, max_val=1.)

def PerceptualLoss(discriminator, flexibility=5e-2, content_weight=6e-4, adversarial_weight=1e-3):
  extractor = tf.keras.Sequential(tf.keras.applications.VGG19(include_top=False, input_shape=(256,256,3)).layers[:11])
  extractor.trainable = False
  discriminator.trainable = False
  def perceptual_loss(y_true, y_pred):
    batch_size = y_true.shape[0]

    true_features = extractor(tf.keras.applications.vgg19.preprocess_input(y_true*255.))
    pred_features = extractor(tf.keras.applications.vgg19.preprocess_input(y_pred*255.))

    mse = tf.reduce_mean(tf.keras.losses.mean_squared_error(true_features, pred_features), axis=[1,2])
    labels = tf.ones(shape=(batch_size,1), dtype=tf.float32) - flexibility * tf.random.uniform(shape=(batch_size, 1))
    bce = tf.keras.losses.binary_crossentropy(labels, discriminator(y_pred))
    return content_weight * mse + adversarial_weight * bce
  return perceptual_loss

def load_models():
  glob_list = glob.glob(os.path.join(MODELS_PATH, "*.h5"))

  if USE_CHECKPOINT and len(glob_list)>=2:
    last_epoch = int(glob_list[-1].split("_")[1].split(".")[0].strip())
    discriminator = tf.keras.models.load_model(os.path.join(MODELS_PATH, f"dis_{last_epoch}.h5"))
    with tf.keras.utils.custom_object_scope({
      'ssim_metric': ssim_metric,
      'psnr_metric': psnr_metric,
      'perceptual_loss': PerceptualLoss(discriminator)
    }):
      generator = tf.keras.models.load_model(os.path.join(MODELS_PATH, f"gen_{last_epoch}.h5"))
  else:
    last_epoch = -1
    discriminator = Discriminator()
    discriminator.compile(
      optimizer=tf.keras.optimizers.Adam(learning_rate=2e-4),
      loss=["binary_crossentropy"]
    )
    generator = Generator()
    generator.compile(
      optimizer=tf.keras.optimizers.Adam(learning_rate=2e-4),
      loss=[PerceptualLoss(discriminator)],
      metrics=[[psnr_metric, ssim_metric]]
    )

  return generator, discriminator, last_epoch+1

def save_models(generator: tf.keras.Model, discriminator: tf.keras.Model, epoch: int):
  generator.save(os.path.join(MODELS_PATH, f"gen_{epoch}.h5"))
  discriminator.save(os.path.join(MODELS_PATH, f"dis_{epoch}.h5"))
  for i in range(epoch):
    if os.path.exists(os.path.join(MODELS_PATH, f"gen_{i}.h5")):
      os.remove(os.path.join(MODELS_PATH, f"gen_{i}.h5"))
    if os.path.exists(os.path.join(MODELS_PATH, f"dis_{i}.h5")):
      os.remove(os.path.join(MODELS_PATH, f"dis_{i}.h5"))
  clean_trash()
````

````python
def save_metrics(gloss, psnr, ssim, mode = "train"):
  filepath = os.path.join(CHECKPOINT_PATH, f"losses_{mode}.chk")
  with open(filepath, "a") as lf:
    line = f"{gloss};{psnr};{ssim}\n"
    lf.write(line)

def load_metrics(mode = "train"):
  filepath = os.path.join(CHECKPOINT_PATH, f"losses_{mode}.chk")
  if not os.path.exists(filepath):
    return None
  glosss = []
  psnrs = []
  ssims = []
  with open(filepath, "r") as lf:
    while (line:=lf.readline()) != None:
      gloss, psnr, ssim = line.split(";")
      glosss.append(gloss)
      psnrs.append(psnr)
      ssims.append(ssim)
  return np.array(glosss, dtype=np.float32),\
    np.array(psnrs, dtype=np.float32),\
    np.array(ssims, dtype=np.float32)
````

````python
generator, discriminator, initial_epoch = load_models()
````

````python
def train_step(epoch, image_loader, last_gloss, last_dloss):
  tqdmloop = tqdm(iterable=iter(image_loader), desc=f"Train Epoch {epoch} => ")
  glosss = [] # overfit check
  psnrs = [] # overfit check
  ssims = [] # overfit check
  dlosses = [] # print only
  gdr = last_gloss/last_dloss
  for idx, (batch_hr, batch_lr) in enumerate(tqdmloop):
    # setup
    batch_size = batch_hr.shape[0]
    batch_sr = generator.predict_on_batch(batch_lr)

    # train discriminator
    discriminator.trainable = True
    if (gdr <= 200):
      real_realness = discriminator.train_on_batch(batch_hr, tf.ones(shape=(batch_size, 1)))
      fake_fakeness = discriminator.train_on_batch(batch_sr, tf.zeros(shape=(batch_size, 1)))

      dlosses.append((real_realness + fake_fakeness)*0.5)
    else:
      real_realness = discriminator.test_on_batch(batch_hr, tf.ones(shape=(batch_size, 1)))
      fake_fakeness = discriminator.test_on_batch(batch_sr, tf.zeros(shape=(batch_size, 1)))
      dlosses.append((real_realness + fake_fakeness)*0.5)

    # train generator
    if (gdr >= 5e-2):
      discriminator.trainable = False
      gloss, psnr, ssim  = generator.train_on_batch([batch_lr], [batch_hr])

      glosss.append(gloss)
      psnrs.append(psnr)
      ssims.append(ssim)
    else:
      gloss, psnr, ssim = generator.test_on_batch([batch_lr], [batch_hr])

      glosss.append(gloss)
      psnrs.append(psnr)
      ssims.append(ssim)

    last_gloss = glosss[-1]
    last_dloss = dlosses[-1]
    gdr = last_gloss/last_dloss


  return np.array(tf.reduce_mean(glosss)),\
    np.array(tf.reduce_mean(psnrs)),\
    np.array(tf.reduce_mean(ssims)),\
    np.array(tf.reduce_mean(dlosses))


def test_step(epoch, image_loader):
  tqdmloop = tqdm(iter(image_loader), desc=f"Test Epoch {epoch} => ")
  last_iter = len(tqdmloop)-1
  glosss = []
  psnrs = []
  ssims = []
  for idx, (batch_hr, batch_lr) in enumerate(tqdmloop):

    gloss, psnr, ssim  = generator.test_on_batch(batch_lr, batch_hr)

    glosss.append(gloss)
    psnrs.append(psnr)
    ssims.append(ssim)

    if idx == last_iter:
      batch_sr = generator.predict_on_batch(batch_lr)
      _, plts = plt.subplots(1,3, figsize=(20,20))
      ind = randint(0, (batch_sr.shape[0])-1)
      plts[0].imshow(np.array(batch_lr[ind]))
      plts[0].set_title("batch_lr")
      plts[1].imshow(np.array(batch_sr[ind]))
      plts[1].set_title("batch_sr")
      plts[2].imshow(np.array(batch_hr[ind]))
      plts[2].set_title("batch_hr")

  return np.array(tf.reduce_mean(glosss)),\
    np.array(tf.reduce_mean(psnrs)),\
    np.array(tf.reduce_mean(ssims))
````

````python
def train(epochs):
  train_loader = ImageLoader(root='/content' if isColab else os.path.curdir, batch_size=8, mode="train")
  test_loader = ImageLoader(root='/content' if isColab else os.path.curdir, batch_size=5, mode="test")

  last_gloss = 1.
  last_dloss = 1.

  for epoch in range(initial_epoch, epochs):
    # train
    gloss_train, psnr_train, ssim_train, dloss_train = train_step(
      epoch,
      train_loader,
      last_gloss=last_gloss,
      last_dloss=last_dloss
    )
    last_gloss = gloss_train
    last_dloss = dloss_train

    # test
    gloss_test, psnr_test, ssim_test = test_step(epoch, test_loader)

    # prints
    print("\nMetrics: gloss\t\tpsnr\t\tssim\t\tdlosses")
    print("Train:\t", "\t".join([
      str(last_gloss),
      str(psnr_train),
      str(ssim_train),
      str(last_dloss),
    ]))
    print("Test:\t", "\t".join([
      str(gloss_test),
      str(psnr_test),
      str(ssim_test),
      "--------"
    ]))
    plt.show()

    # checkpoint
    save_models(generator, discriminator, epoch=epoch)
    save_metrics(last_gloss, psnr_train, ssim_train, mode="train")
    save_metrics(gloss_test, psnr_test, ssim_test, mode="test")
````

````python
train(epochs=135)
````

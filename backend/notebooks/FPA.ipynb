## PREPROCESSING

retrieving the list of cards in all decks retrieved by the matches collection

```python
from pymongo import MongoClient
import json

client = MongoClient("mongodb://localhost:27017")
db = client.DataMining
#version1: db.matches
collection = db.parsed_matches
matchesList = json.loads(json.dumps([card for card in collection.find({})], default=str))
client.close()

cardList = []

for match in matchesList:
  localList = [str(match['card1']['id']),str(match['card2']['id']),str(match['card3']['id']),str(match['card4']['id']),
               str(match['card5']['id']),str(match['card6']['id']),str(match['card7']['id']),str(match['card8']['id'])]
  cardList.append(localList)

print(len(cardList))

for card in cardList:
    print(card)
```

### SAVING A CSV FILE USABLE FROM BACKEND

```python
import pandas as pd

decks_df = pd.DataFrame(cardList)

#version1: decks.csv
decks_df.to_csv("../csv_sources/parsed_decks.csv", index=False, header=False)

decks_df.head(10)
```

## APRIORI PATTERN MINING

using mlxtend library, it requires a conversion to a format with all the possible values in the dataset with a value of true or false for each row.

cardList is filled with the content of the csv file in list of lists format, to be transformed by the transactionEncoder obtaining the necessary table for the apriori algorithm to run

### APRIORI-SPECIFIC PREPROCESSING

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
from csv import reader

cardList = []
#version1: decks.csv
with open("../csv_sources/parsed_decks.csv", "r") as fr:
    csv_content = reader(fr)
    for row in csv_content:
        cardList.append(row)

te = TransactionEncoder()
encoded_dataset = te.fit(cardList).transform(cardList)
df = pd.DataFrame(encoded_dataset, columns=te.columns_)
df.head(10)
```

### APRIORI APPLICATION AND EVALUATION

```python
freq_items = apriori(df, min_support=0.01,use_colnames=True, verbose=2, max_len=8)
freq_items.head(10)
```

### CONFIDENCE METRIC

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(freq_items, metric="confidence")
rules
```

### VISUALIZATION OF CONFIDENCE OF THE RULES

```python
import matplotlib.pyplot as plt
import numpy as np

plt.xlabel("antecedents")
plt.ylabel("confidence")
stringified_antecedents = [str(x) for x in rules["antecedents"]]
plt.scatter(stringified_antecedents, rules['confidence'], alpha=0.5)
```

### VISUALIZATION OF LIFT vs RULES

```python
plt.xlabel("antecedents")
plt.ylabel("lift")
plt.scatter(stringified_antecedents, rules['lift'], alpha=0.5)
```

### VISUALIZATION OF RULE'S ANTECEDENT SUPPORT VS RULE CONFIDENCE

```python
plt.xlabel("antecedent supports")
plt.ylabel("confidence")
plt.scatter(rules['antecedent support'], rules['confidence'], alpha=0.5)
```

### VISUALIZATION OF RULES' ANTECEDENT SUPPORT VS RULE LIFT

```python
plt.xlabel("antecedent supports")
plt.ylabel("lift")
plt.scatter(rules['antecedent support'], rules['lift'], alpha=0.5)
```
